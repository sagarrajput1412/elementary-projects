```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
%matplotlib inline
```

```python
df = pd.read_csv('cartrade-ev.csv')
```

```python
df['review'] = df['reviews'] + ' ' + df['comments']
```

```python
df.drop(['reviews', 'comments'], axis='columns', inplace=True)
```

```python
pd.set_option("display.max_colwidth", 5000)
```

```python
df.head()
```

```python
#create an empty list to collect cleaned review data
cleaned_review =[]
```

```python
for rev in df.review:
    rev = re.sub(r'http\S+|www\S+|https\S+', '', rev, flags=re.MULTILINE)                      # Remove URLs
    rev = re.sub(r'<.*?>', '', rev)                                                          # Remove HTML tags
    rev = re.sub(r'[^A-Za-z0-9\s]', '', rev)                         # Remove non-alphanumeric characters except spaces
    rev = re.sub(r'\s+', ' ', rev).strip()                               # Convert multiple spaces to single space
    rev = re.sub(r'\n', ' ', rev)                                     # Remove newline characters 
    rev = re.sub(r'\d+', '', rev)                                              # Remove digits 
    rev = rev.lower()
    rev = rev.split()
    rev = " ".join(rev)
    cleaned_review.append(rev)
```

```python
df['cleaned_review'] = cleaned_review
```

```python
df.head()
```

```python
df.drop(['review'], axis='columns', inplace=True)
```

```python
df.head()
```

```python
df.rename(columns = {'cleaned_review':'review'}, inplace = True)
```

```python
df.head()
```

```python
df.to_csv('cartrade-ev-vehicle.csv', index = False)
```
