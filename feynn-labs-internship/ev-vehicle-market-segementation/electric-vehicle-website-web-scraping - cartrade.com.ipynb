```python
# time , user
```

```python
for i in soup.find_all('p', class_='user-review__author-days'):
  print(i.text.strip())
```

```python
len(soup.find_all('p',class_='user-review__author-days'))
```

```python
data = []

for i in soup.find_all('p', class_='user-review__author-days'):
    text = i.get_text()
    # Split the text into 'time' and 'user' parts
    time, user = map(str.strip, text.split('|'))
    data.append([time, user])
    
```

```python
data
```

```python
divs = soup.find_all('p', class_='user-review__author-days')

# Initialize lists for time and user
time_of_review = []
user_name = []

# Extract and split the data
for div in divs:
    text = div.get_text().strip()
    time, user = map(str.strip, text.split('|'))
    time_of_review.append(time)
    user_name.append(user)

# Display the lists
# print("time:", time_of_review)
# print("user:", user_name)
```
